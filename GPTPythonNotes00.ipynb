Basic syntax:
- programs are written line by line,
- concatenate strings into one logical character \ (+\),
- two logical strings are combined into one physical string with the symbol ;,
- objects in brackets {},[],() are transferred to the next lines after commas without \,
- lines after # or inside """ """ - comments are not perceived as programmatic,
- lines end without any characters except for nested constructs,
- nested blocks are formed with a colon : and indents (usually 4).

Sintaxis básica:
- los programas se escriben línea por línea,
- concatenar cadenas en un carácter lógico \ (+\),
- dos cadenas lógicas se combinan en una cadena física con el símbolo;,
- los objetos entre paréntesis {},[],() se transfieren a las líneas siguientes después de la coma sin \,
- líneas después de # o dentro de """ """ - los comentarios no se perciben como programáticos,
- las líneas terminan sin ningún carácter excepto las construcciones anidadas,
- los bloques anidados se forman con dos puntos: y sangrías (normalmente 4).
