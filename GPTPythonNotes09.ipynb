```python
# custom functions / funciones personalizadas
def table_print(table):
    """
    The function prints a table line by line,
    taking a two-dimensional list as an argument.
    La función imprime la tabla línea por línea,
    tomando una lista bidimensional como un argumento.

    """
    col_number = max(map(len, table))
    flatten = list(map(str, sum(table, [])))
    col_width = max(map(len, flatten)) + 5
    line = (col_number * col_width + 1) * '-'
    line = "\033[1;38;5;177m" + line + " \033[0;0m"
    sep = "\033[1;38;5;177m|\033[0;0m"
    for row in table:
        print(line)
        row_str = sep
        for r in row:
            row_str += str(r).center(col_width - 1) + sep
        row_str += (col_number - len(row)) * (' '.center(col_width - 1) + sep)
        print(row_str)
    print(line)
    return
```

```python
# the custom function to compare operators
# la función personalizada para comparar operadores
import time, inspect, sys
from io import StringIO
def faster_on_average(f1, f2, data, iters=5):
    lst, n = data; t= []
    old_stdout = sys.stdout
    sys.stdout = my_stdout = StringIO()
    my_stdout.write(inspect.getsource(f1))
    my_stdout.write(inspect.getsource(f2))
    sys.stdout = old_stdout
    for i in range(iters):
        lst1, lst2 = lst.copy(), lst.copy()
        start = time.time()
        f1(lst1, n)
        t1 = time.time() - start
        start = time.time()
        f2(lst2, n)
        t2 = time.time() - start
        t.append(t1 / t2)
    f1_str, f2_str, _ = my_stdout.getvalue().split('\n')
    result = f2_str + '\n'
    result += f"is {sum(t) / iters} times faster than \n"
    result += f1_str
    return result
```

```python
data = (list(range(10**6)), 2)
def f1(lst, n): lst = lst * n; return lst
def f2(lst, n): lst *= n; return lst
print(faster_on_average(f1, f2, data, iters=10))
```

`yield` vs `return`

```python
import sys
# list initiator / iniciador de lista
def n_list(n):
    number, numbers = 0, []
    while number < n:
        numbers.append(number)
        number += 1
    # it stops calculations
    # detiene los cálculos
    return numbers
n_sum = sum(n_list(10e6))
print(n_sum, '=>', sys.getsizeof(n_list(10e6)))
# sequence generator / generador de secuencia
def n_gen(n):
    num = 0
    while num < n:
        # it pauses calculations until next() will called
        # pausa los cálculos hasta que se llame a next()
        yield num
        num += 1
n_sum = sum(n_gen(10e6))
print(n_sum, '=>', sys.getsizeof(n_gen(10e6)))
```

```python
# CODEJAM QUALIFICATION ROUND 2022 TASK 1
# аргументы-ключи (ключевые слова)
# произвольный порядок при вызове
def gen_table(symbol, rows, columns):
    for i in range(2 * rows + 1):
        for j in range(2 * columns + 1):
            if (i < 2 and j < 2):
                yield symbol + ' '
            else:
                el = ((j + 1) % 2) * ((i + 1) % 2) * '+'
                el += (j % 2) * ((i + 1) % 2) * '-:-'
                el += ((j + 1) % 2) * (i % 2) * '|'
                el += (j % 2) * (i % 2) * (' ' + symbol + ' ')
                yield el
        yield '\n'
gen = gen_table(rows=4, columns=7, symbol='α')
print(''.join(gen))
```

Classification of functions by scope

Clasificación de funciones por alcance
- con argumentos locales (declarados internamente)
- con los globales (declarados externamente)
  - las funciones pueden cambiar variables globales

```python
arr = [1, 2, 3, 4, 5]
def create_arr(n):
    arr = list(range(1, n+1))
    print(arr)
def display_arr():
    global arr
    print(arr)
def transform_arr(n):
    global arr
    if n > len(arr):
        arr = list(range(1, n+1))
    print(arr)
def display_transform_arr(m, n):
    arr = list(range(1, n+1))
    def transform_arr():
        nonlocal arr, m, n
        if m > n:
            arr = list(range(n, m+1))
        return arr
    arr = transform_arr()
    print(arr)
```

```python
# the local argument arr does not affect the global one
# el argumento local arr no afecta al global
create_arr(7), arr
```

```python
# the global argument arr was used but not changed
# se utilizó el argumento global arr pero no se modificó
display_arr(), arr
```

```python
# the global argument arr has changed
# el argumento global arr ha cambiado
transform_arr(9), arr
```

```python
# a decorator-function modifies other functions
# una función decoradora modifica otras funciones
def deco(func):
    def function_wrapper(x):
        print(f"now function `{func.__name__}` will be called")
        func(x)
        print(f"function `{func.__name__}` was changed after decorator `deco`")
    return function_wrapper
def f(x):
     print(f"function `{f.__name__}` works with argument `{str(x)}`")
print(f"Call function `{f.__name__}` without decorator `deco`")
f("abc"); print(50 * '=' + '\n')
print(f"Call function `{f.__name__}` by decorator `deco`")
f = deco(f) ; print(50 * '=' + '\n')
print(f"Call function `{f.__name__}` after decorator `deco`")
f("123")
```

```python
# simple universal decorator
# un decorador universal simple
def upper_deco(f):
    def inf(*arg, **kwarg):
        return f(*arg, **kwarg).upper()
    return inf
@upper_deco
def example1(name):
    return f'hello, {name}'
@upper_deco
def example2(x, y):
    return f'{x = }\n{y = }\nx + y = {x + y}'
print(example1('tim'), example2(2, 3), sep='\n')
```

```python
# information about a running time
# información sobre un tiempo de ejecución
import functools, time
def timer(f):
    @functools.wraps(f)
    def wtimer(*args, **kwargs):
        start = time.perf_counter()
        value = f(*args, **kwargs)
        end = time.perf_counter()
        string = f"the running time of the function {f.__name__!r}: "
        string += f"{(end - start):.4f} seconds"
        print(string)
        return value
    return wtimer
@timer
def example5(n):
    s = 0
    for _ in range(n):
        s += sum([i**2 for i in range(n)])
    return s
example5(10**3)
```

```python
import functools
# a decorator for call counting
# un decorador para el conteo de llamadas
class NumCalls:
    def __init__(self, f):
        functools.update_wrapper(self, f)
        self.f = f
        self.ncalls = 0
    def __call__(self, *args, **kwargs):
        self.ncalls += 1
        print(f"number of calls {self.f.__name__!r}: {self.ncalls}")
        return self.f(*args, **kwargs)
# a decorator for repeating
# un decorador para repetir
def repeat(_f=None, *, n=2):
    def repeat_deco(f):
        @functools.wraps(f)
        def wrepeat(*args, **kwargs):
            value = "the function was not called"
            for _ in range(n):
                value = f(*args, **kwargs)
            return value
        return wrepeat
    if _f is None:
        return repeat_deco
    else:
        return repeat_deco(_f)
@repeat
@NumCalls
def example6(): print('hello')
example6()
print(35 * "=")
@repeat(n=4)
@NumCalls
def example7(): print('hello')
example7()
```

```python
def is_palindrome(string):
    if len(string) <= 1:
        return True
    else:
        check = string[0] == string[-1]
        return check and is_palindrome(string[1:-1])
is_palindrome('1235321'), is_palindrome('123321')
```

```python
# a decorator to count the number of function calls
# un decorador para contar el número de llamadas a funciones
def call_count(func):
    def count(*args, **kwargs):
        count.calls += 1
        return func(*args, **kwargs)
    count.calls = 0
    return count
# a recursive generator-function with call counting
# una función generadora recursiva con conteo de llamadas
@call_count
def recursive_pow(lst,n):
    if lst:
        yield lst[0]**n
        yield from recursive_pow(lst[1:], n)
rg = recursive_pow(range(10), 5)
for i in range(7):
    print(next(rg), end=' ')
print()
# the number of calls / el número de llamadas
print(f"{recursive_pow.calls = }", '\n'+65*'-')
# the remaining values in the generator
# los valores restantes en el generador
print(list(rg))
# calls to the built-in list() function are also counted
# también se cuentan las llamadas a la función incorporada list()
print(f"{recursive_pow.calls = }", '\n'+65*'-')
# create a new generator with the same function
# crear un nuevo generador con la misma función
rg = recursive_pow(range(5), 3)
for i in range(5):
    print(next(rg), end=' ')
print()
# the counting continues / el conteo continúa
print(f"{recursive_pow.calls = }")
```

```python
# sometimes recursion to the previous value is not a fantastic idea
# a veces la recursión al valor anterior no es una idea fantástica
import time
def recursive_power(base, pow):
    if (base == 1) or (pow == 0):
        return 1
    else:
        return base * recursive_power(base, pow-1)
start = time.perf_counter()
value = recursive_power(9, 900)
end = time.perf_counter()
string = f"the running time of the function recursive_power(): "
string += f"{(end - start):.4f} seconds"
print(string)
def fast_power(base, pow):
    result = 1
    while pow:
        if pow & 1:
            result *= base
        pow >>= 1
        base *= base
    return result
start = time.perf_counter()
value = fast_power(9, 10**3)
end = time.perf_counter()
string = f"the running time of the function fast_power(): "
string += f"{(end - start):.4f} seconds"
print(string)
```

```python
# some classic recursions
# algunas recursiones clásicas
count_merge_sort, count_merge = 0, 0
def merge(left, right):
    global count_merge
    count_merge += 1
    print(f"{count_merge = }")
    print(f"{left = }, {right = }")
    left_index, right_index = 0, 0
    result = []
    while left_index < len(left) and right_index < len(right):
        if left[left_index] < right[right_index]:
            result.append(left[left_index])
            left_index += 1
        else:
            result.append(right[right_index])
            right_index += 1
    result += left[left_index:]
    result += right[right_index:]
    print(f"{result = }")
    return result
def merge_sort(array):
    global count_merge_sort
    count_merge_sort += 1
    print(f"{count_merge_sort = }")
    if len(array) <= 1:
        return array
    half = len(array) // 2
    left = merge_sort(array[:half])
    right = merge_sort(array[half:])
    return merge(left, right)
merge_sort([1, -5, 2, 3, 5, -7, 8])
```

```python
# the simplest encoder/decryptor
# el codificador/descifrador más simple
def encrypt(string):
    crypt = lambda c, i: chr(ord(c)+8300+round(1000*(i+1)**.1))
    s = map(crypt, string, range(len(string)))
    return ''.join(s)
def decrypt(string):
    crypt = lambda c, i: chr(ord(c)-8300-round(1000*(i+1)**.1))
    s = map(crypt, string, range(len(string)))
    return ''.join(s)
s = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
encrypt(s), len(set(encrypt(s))), decrypt(encrypt(s)), len(set(s))
```

```python
# lambda as a sort key
# lambda como clave de clasificación
lst = [-1, -5, -2, 3, 1, 7, -3, 4, -8]
sorted_lst_abs = sorted(lst, key=lambda x: abs(x))
print(sorted_lst_abs)
```

```python
# to apply a function multiple times to collections
# with a “cumulative” total according to the lambda law
# aplicar una función varias veces a las colecciones
# con un total “acumulado” según la ley lambda
from functools import reduce
lst = list(range(1, 11))
lst_sum = reduce(lambda x, y: x + y, lst)
print(lst_sum)
lst_prod = reduce(lambda x, y: x * y, lst)
print(lst_prod)
```

Functions of a single stroke, just for fun
- anonymous functions, lambda expressions, or function literals (the same things)

Funciones de un solo golpe, sólo por diversión
- funciones anónimas, expresiones lambda o literales de funciones (lo mismo)

```python
def sr1(x): return str(5 * x) + '!'
sr2 = lambda x: str(5 * x) + '!'
sr1(3), sr1('hi'), sr2(1), sr2('hi')
```

```python
sum_digits = lambda n: n if n < 10 else (n % 10) + sum_digits(n // 10)
sum_digits(12345654321)
```

```python
# the factorial recursion / la recursividad factorial
fact = lambda n: 1 if n <= 1 else n * fact(n - 1)
fact(10)
```

```python
triangular_number = lambda n: 0 if n == 0 else n + triangular_number(n-1)
for i in range(20):
    print(triangular_number(i), end=' ')
```

```python
# x^n
exponentiation = lambda x, n: 1 if n == 0 else x * exponentiation(x, n-1)
exponentiation(3, 101)
```

```python
# greatest common divisor / máximo común divisor
gcd = lambda a, b: a if b == 0 else gcd(b, a % b)
gcd(120, 25)
```

```python
# a number in binary / un número en binario
decimal_to_bin = lambda n: '0' if n == 0 else decimal_to_bin(n // 2) + str(n % 2)
decimal_to_bin(100001)
```

```python
bin_to_decimal = lambda n: n if n < 2 else (n % 10) + 2 * bin_to_decimal(n // 10)
bin_to_decimal(11000011010100001)
```

```python
# reverse a string / invertir una cadena
reverse_str = lambda s: s if len(s) <= 1 else reverse_str(s[1:]) + s[0]
reverse_str('12345')
```
