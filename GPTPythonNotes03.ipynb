```python
# solve number puzzles
# resolver acertijos num√©ricos
def get_eq(s, strings):
    s = s.replace(" ", "")
    letter = ''
    for i in s:
        cond1 = i.isdigit()
        cond2 = i in ['+', '-', '*', '/', '=']
        if not (cond1 or cond2):
            letter = i
            break
    if (not letter):
        try: eq = eval(s.replace("=", "=="))
        except: return
        if eq:
            strings.append(s.replace("+", " + ").replace("=", " = "))
        return
    for n in range(10):
        n = str(n)
        if (s.find(n) > -1): continue
        get_eq(s.replace(letter, n), strings)
s1, strings1 = "SEND + MORE = MONEY", []
s2, strings2 = "$ + $ + $ + $ = @", []
get_eq(s1, strings1)
print(*strings1, sep='\n')
get_eq(s2, strings2)
print(*strings2, sep='\n')
```

```python
def gen_table(r,c):
    for i in range(2*r+1):
        for j in range(2*c+1):
            if (i<2 and j<2):
                yield '.'
            else:
                el = ((j + 1) % 2) * ((i + 1) % 2) * '+' + \
                     (j % 2) * ((i + 1) % 2) * '-' + \
                     ((j + 1) % 2) * (i % 2) * '|' + \
                     (j % 2) * (i % 2) * '.'
                yield el
        yield '\n'
# T = int(input())
T = 1
for t in range(T):
#     N = input()
    N = '3 5'
    print('Case #{}:'.format(t+1))
    [R, C] = [int(n) for n in N.split()]
    gen = gen_table(R,C)
    print(''.join(gen))
```

 String arts and games

 Artes de cuerdas  y juegos

```python
import random
rock = """
    _______
---'   ____)
      (_____)
rock  (_____)
      (____)
---.__(___)
"""
paper = """
     _______
---'    ____)____
           ______)
paper      _______)
         _______)
---.__________)
"""
scissors = """
    _______
---'   ____)_____
         ________)
scissors _________)
       (____)
---.___(___)
"""
player1 = random.choice([rock, paper, scissors])
player2 = random.choice([rock,paper,scissors])
print("#1: \n", player1)
print("#2: \n", player2)
if player1 == rock and player2 == paper:
    print("the 2nd player wins")
elif player1 == paper and player2 == scissors:
    print("the 2nd player wins")
elif player1 == scissors and player2 == rock:
    print("the 2nd player wins")
elif player1 == player2:
    print("draw")
else:
    print("the 1nd player wins")
```
