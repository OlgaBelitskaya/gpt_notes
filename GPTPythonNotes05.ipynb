Knapsack problems like
- total weight `w`
- list of value indicators (weights of value) `wlst`
- list of values `plst`

the value set: w = 10 => max sum for these values?

Problemas de mochila como
- peso total `w`
- lista de indicadores de valor (pesos de valor) `wlst`
- lista de valores `plst`

el conjunto de valores: w = 10 => suma máxima para estos valores?

```python
# dynamically updated lists
# listas actualizadas dinámicamente
def max_pw(w, n, wlst, plst, pw):
    if n == 0 or w == 0:
        return 0
    elif pw[n][w] is not None:
        return pw[n][w]
    elif wlst[n-1] <= w:
        pw[n][w] = max(
            plst[n-1] + max_pw(w - wlst[n-1], n-1, wlst, plst, pw),
            max_pw(w, n-1, wlst, plst, pw))
        return pw[n][w]
    else:
        pw[n][w] = max_pw(w, n-1, wlst, plst, pw)
        return pw[n][w]
w1 = 12
wlst1 = [1, 2, 3, 6, 7, 8, 9]
plst1 = [2, 6, 18, 32, 22, 27, 41]
n1 = len(plst1)
pw1 = [[None] * (w1+1) for _ in range(n1+1)]
max_pw1 = max_pw(w1, n1, wlst1, plst1, pw1)
print(*pw1, max_pw1, sep='\n')
```

tuples vs lists
- useful for objects connected to each other
- suitable for heterogeneous (different) data types
- immutable, so iterating is a bit faster
- can be used as keys for a dictionary
- data objects, implemented as tuples, are more protected against overwriting

tuplas vs listas
- útil para objetos conectados entre sí
- adecuado para tipos de datos heterogéneos (diferentes)
- inmutable, por lo que la iteración es un poco más rápida
- se puede utilizar como claves para un diccionario
- los objetos de datos, implementados como tuplas, están más protegidos contra la sobrescritura
